```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df18= pd.read_csv("data_raw/divorcios_2018.csv")
df19= pd.read_csv("data_raw/divorcios-2019.csv")
df20= pd.read_csv("data_raw/divorcios-2020.csv")
df21= pd.read_csv("data_raw/divorcios-2021.csv")
df22= pd.read_csv("data_raw/divorcios-2022.csv")
df23= pd.read_csv("data_raw/divorcios-2023.csv")
df24 = pd.read_csv("data_raw/divorcios-2024.csv")
df25 = pd.read_csv("data_raw/divorcios-2025.csv")

```

```python
df18 = df18.rename(columns={
    'fecha_creacion_nr_divorcio': 'fecha_divorcio',
    'genero_1':'sexo_1',
    'genero_2':'sexo_2'
})

df19 = df19.rename(columns={
    'fecha_creacion_divorcio': 'fecha_divorcio',
    'genero_1':'sexo_1',
    'genero_2':'sexo_2'
})
```

```python
lista_df = [df18, df19, df20, df21, df22, df23, df24, df25]

df = pd.concat(lista_df, ignore_index=True)
```

```python
df['fecha_divorcio'] = df['fecha_divorcio'].str.replace('"', '', regex=False)
df['fecha_matrimonio'] = df['fecha_matrimonio'].str.replace('"', '', regex=False)
```

```python
df['fecha_matrimonio'] = df['fecha_matrimonio'].replace('0197-01-01', '1997-01-01', regex=False)
```

```python
df['fecha_matrimonio'] = df['fecha_matrimonio'].replace('1760-01-01', '1960-01-01', regex=False)
```

```python
df['fecha_divorcio'] = pd.to_datetime(df['fecha_divorcio'], format='mixed')
df['fecha_matrimonio'] = pd.to_datetime(df['fecha_matrimonio'], format='mixed')
```

```python
df['fecha_divorcio'] = pd.to_datetime(df['fecha_divorcio']).dt.date
df['fecha_matrimonio'] = pd.to_datetime(df['fecha_matrimonio']).dt.date
```

```python
df['tiempo_transcurrido'] = df['fecha_divorcio'] - df['fecha_matrimonio']
```

```python
df['tiempo_transcurrido'] = pd.to_timedelta(df['tiempo_transcurrido'])

df['anios_matrimonio'] = (df['tiempo_transcurrido'].dt.days / 365.25).round(1)
```

```python
df['fecha_divorcio'] = pd.to_datetime(df['fecha_divorcio'])
df['fecha_matrimonio'] = pd.to_datetime(df['fecha_matrimonio'])

df['anio_divorcio'] = df['fecha_divorcio'].dt.year
df['anio_matrimonio'] = df['fecha_matrimonio'].dt.year
```

```python
df_anios = df.groupby('anio_divorcio').size().reset_index(name='total_divorcios')
```

```python
df['mes_divorcio'] = df['fecha_divorcio'].dt.month
```

```python
df_meses = df.groupby('mes_divorcio').size().reset_index(name='total_divorcios')
```

```python
df['sexo_1'] = df['sexo_1'].replace('NO DECLARA', 'No declara', regex=False)
df['sexo_1'] = df['sexo_1'].replace('DESCONOCIDO', 'No declara', regex=False)
df['sexo_1'] = df['sexo_1'].replace('Indefinido', 'No declara', regex=False)
df['sexo_1'] = df['sexo_1'].replace('INDEFINIDO', 'No declara', regex=False)
df['sexo_1'] = df['sexo_1'].replace('DESCONOCIDO"', 'No declara', regex=False)
df['sexo_1'] = df['sexo_1'].replace('INDEFINIDO"', 'No declara', regex=False)
df['sexo_1'] = df['sexo_1'].replace('NO DECLARA"', 'No declara', regex=False)
df['sexo_1'] = df['sexo_1'].replace('F"', 'Femenino', regex=False)
df['sexo_1'] = df['sexo_1'].replace('F', 'Femenino', regex=False)
df['sexo_1'] = df['sexo_1'].replace('M"', 'Masculino', regex=False)
df['sexo_1'] = df['sexo_1'].replace('M', 'Masculino', regex=False)
```

```python
def determinar_tipo_pareja(fila):

    if (fila['sexo_1'] == "No declara" or pd.isna(fila['sexo_1']) or
        fila['sexo_2'] == "No declara" or pd.isna(fila['sexo_2'])):
        return 'desconocido'

    elif fila['sexo_1'] == fila['sexo_2']:
        return 'mismo sexo'

    else:
        return 'dif sexo'

df['tipo_pareja'] = df.apply(determinar_tipo_pareja, axis=1)
```

```python
df_tp = df.groupby('tipo_pareja').size().reset_index(name='total_divorcios')
```
