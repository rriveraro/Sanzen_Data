```python
plt.figure(figsize=(9, 4))
sns.kdeplot(data=df[df['anios_matrimonio'] <= 15], x='anios_matrimonio', fill=True, common_norm=False, palette='crest')
plt.title('Picos de Riesgo: ¿A qué año exacto se quiebran más los matrimonios?')
plt.show()
```
